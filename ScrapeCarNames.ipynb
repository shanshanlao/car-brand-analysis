```python
import requests
import time
from bs4 import BeautifulSoup
```

```python
# User-defined function for Scraping the data

def getData(url):
    page = requests.get(url, headers = {'My-Header':'value'})
    return page.text
```

```python
# Pass the urll to get the html data

url = "https://en.wikipedia.org/wiki/List_of_automobile_manufacturers"

htmldata = getData(url)
soup = BeautifulSoup(htmldata, "html.parser")

print(soup)
```

```python
# Find all <a> tags within <li> tag

manufacturers_tags = soup.find_all("li")

manufacturers = []
for tag in manufacturers_tags:
    a_tag = tag.find("a")
    if a_tag:
        if a_tag.text == 'Vinaxuki':
            break
        manufacturers.append(a_tag.text)

manufacturers.append('Vinaxuki')    
manufacturers.remove('A')

# Print the names of the manufacturers
print(len(manufacturers))
```

```python
import pandas as pd

df = pd.DataFrame(manufacturers, columns = ['Make'])
df.to_csv("car_companies.csv")
```
